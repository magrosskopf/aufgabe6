```python
import pandas as pd
csv = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv', error_bad_lines=False)
#Leere Zeilen entfernen, weil keine Inhalte enthalten sind
csv.dropna(how="all", inplace=True) 
#Die Spalten ID und full_name werden aus Gründen der Datensparsamkeit entfernt, da jede Zeile einen Index durch pandas hat und die ID überflüssig macht und der gesamte Name jeder Zeit aus den Spalten first_name und last_name generiert werden kann.
csv.drop(['id', 'full_name'], inplace=True, axis=1)
# Zeilen die identisch sind, werden auf eine reduziert
csv.drop_duplicates(inplace=True)
#Altersangaben werden zuerst von strings zu Floats umgewandlet. Gibt es beim umwandelen eine Fehler, z.B. weil der String keine Zahl enthält sondern das Wort "old", wird die Zeile entfernt. Für Old kann kein Standardwert gesetzt werden, da dieser Begriff eine große Reihe an Zahlen abdeckt.
#Alle Floats werden zu einem positiven Float umgewandelt, damit Tippfehler wie -78 zu 78 umgewandelt werden.
csv.age = pd.to_numeric(csv.age, errors='coerce').abs()
# Zeilen in denen kein Geschlecht angegeben wird, wird ein Standardwert (hier "not specified") gesetzt.
csv["gender"].fillna("not specified", inplace = True)
#  Restliche Zeilen die NaN enthalten entfernen, da die nicht korrigierten Datensätze unvollständig sind
csv.dropna(inplace=True)
# Index neu vergeben
csv.reset_index(inplace=True, drop=True)
csv
```
